#### Import necessary libraries

```python
# for scraping 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup as bs
from bs4 import Tag, NavigableString
import xlsxwriter

# misc
import re as re # regex 
import time
import pandas as pd

# accessing config file
# import configparser # for getting values of a config file, config file is where you store values just like an env 
# config = configparser.ConfigParser()
# config.read_file(open('../establishment.cfg')) # read the content of this config file

# accesing env file 
import os 
from dotenv import load_dotenv # to access the secret keys we've hidden in a separate file 
load_dotenv() # grab values inside env file
```

#### SSD Setup

```python
PATH = os.getenv("WEBDRIVER_PATH")
```

```python
# initialize web driver that would control the web browser
ser = Service(PATH)
op = webdriver.ChromeOptions()
driver = webdriver.Chrome(service=ser, options=op)

# website we wanted to access 
driver.get("https://southstardrug.com.ph/pages/store-locations")
```

#### Web Scraping Process

```python
# Creating lists
branch_name = []
branch_address_storeHours_phone_removeHrTag = []

def ssd_branches_scrape(): 
    # extract source code 
    store_location_page = driver.page_source

    # read extracted soure code
    linkedin_soup = bs(store_location_page.encode("utf-8"), "html")
    linkedin_soup.prettify()
    regions = linkedin_soup.findAll("div",{"class":"content"})
    
    for branch in regions:
        # get branch name 
        for hr in branch.findAll('hr'):
            next_hr = hr.nextSibling
            branch_name_value = str(next_hr).strip() # understand what this code doe
            if (branch_name_value != True):
                branch_name.append(branch_name_value)
            else:
                branch_name_value = 0
                branch_name.append(branch_name_value) 
            
        # get branch address, phone number, and store hours
        for br in branch.findAll('br'):
            next_br = br.nextSibling # grab the next navigableString or tag, in this case it's the string which contains branch name
            branch_address_value = str(next_br).strip() # understand what this code does 
            if (branch_address_value != True):
                branch_address_storeHours_phone_removeHrTag.append(branch_address_value)
            else:
                branch_address_value = 0
                branch_address_storeHours_phone_removeHrTag.append(branch_address_value) 
```

#### Execute scraping

```python
ssd_branches_scrape()
```

#### Storing scraped data into specified file format

```python
data_name = {
    "Name": branch_name
    
}

data_address_storesHours_phone_removeHrTag = {
    "Address_StoreHours_Phone_RemoveHrTag": branch_address_storeHours_phone_removeHrTag
}

# save to csv 
# df = pd.DataFrame(data)
# df.head(5)
# df.to_csv("user_posts.csv", encoding='utf-8', index=False)

# save to excel 
df1 = pd.DataFrame(data_name)
writer1 = pd.ExcelWriter("../data/ssd-name.xlsx", engine='xlsxwriter')
df1.to_excel(writer1, index =False)
writer1.save()

df2 = pd.DataFrame(data_address_storesHours_phone_removeHrTag)
writer2 = pd.ExcelWriter("../data/ssd-address_storeHours_phone_removeHrTag.xlsx", engine='xlsxwriter')
df2.to_excel(writer2, index =False)
writer2.save()
```
